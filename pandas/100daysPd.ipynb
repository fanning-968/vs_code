##### 通过字典创建series

```python
ser2 = pd.Series({'一季度': 320, '二季度': 180, '三季度': 300, '四季度': 405})
ser2
```

```python
print(ser2[0], ser2[1], ser2[2], ser2[3])
ser2[0], ser2[3] = 350, 360
print(ser2)
```

```python
print(ser2['一季度'], ser2['三季度'])
ser2['一季度'] = 380
print(ser2)
```

```python
print(ser2[1:3])
print(ser2['二季度':'四季度'])
```

```python
print(ser2[['二季度', '四季度']])
print(ser2[[1,3]])
ser2[['二季度', '四季度']] = 500, 520
print(ser2)
```

```python
print(ser2[ser2 >= 500])
```

```python
# 求和
print(ser2.sum())
# 求均值
print(ser2.mean())
# 求最大
print(ser2.max())
# 求最小
print(ser2.min())
# 计数
print(ser2.count())
# 求标准差
print(ser2.std())
# 求方差
print(ser2.var())
# 求中位数
print(ser2.median())
```
