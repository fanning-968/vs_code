```python
import numpy as np 
import pandas as pd 

# create a dataframe with 3 columns and 5 rows      
data = np.random.randint(0, 100, (5, 3))
index = pd.date_range('2025-01-01', periods=5, freq='W')
name = ['zhangsan', 'lisi', 'wangwu', 'zhaoliu', 'chenfei']
df = pd.DataFrame(data, index=index, columns=['语文', '数学', '英语'])
df['姓名'] = name
df.index.name = '日期'
new_order = ['姓名', '语文', '数学', '英语']
df = df[new_order]
# df[(df.语文 > 80) & (df.数学 > 40)]

# df[(df['语文'] > 60) & (df['数学'] > 40)]

df['zongfen'] = df['语文'] + df['数学'] + df['英语']
df['chonfu'] = df.duplicated('英语')
df.sort_values(by='zongfen', ascending=False)

```
